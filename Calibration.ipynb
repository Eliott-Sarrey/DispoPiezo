Write in the window the value of the input voltage in V and press enter to validate. Repeat with different values from 0 to 20V. Write 'done' and press enter to end the measurements.

Those factors have to be written in the code bellow the section "PRESSURE AND VOLTAGE MEASUREMENTS" in the file "Measurements.ipynb"

# Calibration factor for input 1

```python
%run ./src/Modules.ipynb  

RealVArr = []
VArr = []

measureDuration=1


while 1:
    inP = ''
    while inP.replace('e', '').replace('E', '').replace('.', '').isdigit() == False and inP != 'done':
        inP = input('Input voltage [V]')
        if inP.replace('e', '').replace('E', '').replace('.', '').isdigit() == False and inP != 'done':
            print('Please use the following format: 100 or 1e2')
    if inP == 'done': break
    RealV = float(inP)

    voltage, pressure   = 1, 1 #getRFInputs(measureDuration)                              

    RealVArr.append(RealV)
    VArr.append(voltage)

RealVArr, VArr = np.array(RealVArr), np.array(VArr)



VRatioData = graph(RealVArr, RealVArr/VArr, 'Real V [V]', 'Real voltage divided by measured voltage [1]', 'Calibration ratio')
VMean = graph(RealVArr, np.mean(RealVArr/VArr)*RealVArr/RealVArr, symbol='--', color='C1')
plotGraph([VRatioData, VMean], None, 'VsResistance', points = True, save=False)
print('Mean value: '+str(np.mean(RealVArr/VArr)))
```
